# Importing Data from JSON files

**Method 1: Using open and json.load:**

```python
#Step - 1 ( import pandas and json)
import pandas as pd
import json
```

```python
# Step 2: Read the JSON file using pandas
# Replace 'your_file.json' with the actual path to your JSON file
with open('blockbusters.json', 'r') as json_file:
    data = json.load(json_file)
```

```python
# Step 3: Create a DataFrame from the JSON data
df = pd.DataFrame(data)
df
```

**Method 2: Directly loading a JSON file with pd.read_json:**

```python
df1 = pd.read_json('blockbusters.json')
```

**To work with nested lists(genres, belongs_to_collection) in the JSON and convert them into a DataFrame:**

```python
# belongs_to_collection is separeted in id, name and poster_path and backdrop_path but genres not separated
pd.json_normalize(data = data, sep='_')
```

```python
# for genres use parameter record path 
df2 = pd.json_normalize(data=data, record_path="genres")
df2
```

```python
#but from this we not understand for which movie the genre is, use parameter meta and record_prefix
df3 = pd.json_normalize(data=data, record_path='genres', meta=['title', 'id'], record_prefix='genres')
df3
```
